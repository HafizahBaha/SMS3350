## Equation of an Ellipse



The graph of
$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$
is an **ellipse** where

*   $a$ represents ..., and 
*   $b$ represents ...


--------------------------------------------------------------

### Let's play!

```python
from ipywidgets import interact

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()

x = 0
y = 0
a = 1
b = 1

p = figure(title="Let's play!", plot_height=450, plot_width=450, x_range=(-5,5), y_range=(-5,5),
           background_fill_color='#efefef')

def update(a=1, b=1):
    r = p.ellipse(x, y,
              width = a,
              height = b)
    push_notebook()

show(p, notebook_handle=True)
interact(update, a=(1,10), b=(1,10))
```
